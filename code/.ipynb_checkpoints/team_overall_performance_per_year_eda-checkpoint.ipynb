```python
import pandas as pd
import numpy as np
import props as p
```

```python
# Overall NFL regular season offense performance by team 2010 - 2018.
o_reg_18 = pd.read_csv('../data/top_nfl_teams/regular_season/offense/overall/overall_18.csv')
o_reg_17 = pd.read_csv('../data/top_nfl_teams/regular_season/offense/overall/overall_17.csv')
o_reg_16 = pd.read_csv('../data/top_nfl_teams/regular_season/offense/overall/overall_16.csv')
o_reg_15 = pd.read_csv('../data/top_nfl_teams/regular_season/offense/overall/overall_15.csv')
```

```python
# Overall NFL regular season defense performance by team 2010 - 2018.
d_reg_18 = pd.read_csv('../data/top_nfl_teams/regular_season/defense/overall/overall_18.csv')
d_reg_17 = pd.read_csv('../data/top_nfl_teams/regular_season/defense/overall/overall_17.csv')
d_reg_16 = pd.read_csv('../data/top_nfl_teams/regular_season/defense/overall/overall_16.csv')
d_reg_15 = pd.read_csv('../data/top_nfl_teams/regular_season/defense/overall/overall_15.csv')
```

```python
o_reg_18.head()
```

```python
# Add a new column to represent the year associated with each regular season.

# Regular season offense dataframes.
o_reg_18['Year'] = 2018
o_reg_17['Year'] = 2017
o_reg_16['Year'] = 2016
o_reg_15['Year'] = 2015

# Regular season defense dataframes.
d_reg_18['Year'] = 2018
d_reg_17['Year'] = 2017
d_reg_16['Year'] = 2016
d_reg_15['Year'] = 2015
```

```python
# Assignment of names to all dataframes.

# Regular season offense dataframes:
o_reg_18.name = 'Regular Season Offense Performance by Team 2018'
o_reg_17.name = 'Regular Season Offense Performance by Team 2017'
o_reg_16.name = 'Regular Season Offense Performance by Team 2016'
o_reg_15.name = 'Regular Season Offense Performance by Team 2015'

# Regular season defense dataframes:
d_reg_18.name = 'Regular Season Defense Performance by Team 2018'
d_reg_17.name = 'Regular Season Defense Performance by Team 2017'
d_reg_16.name = 'Regular Season Defense Performance by Team 2016'
d_reg_15.name = 'Regular Season Defense Performance by Team 2015'
```

```python
# Add all regular season offense and defense dataframes to a list to allow use of print_props funtion.

# Regular season offense:
o_reg_dfs = [o_reg_18, o_reg_17, o_reg_16, o_reg_15]

# Regular season defense:
d_reg_dfs = [d_reg_18, d_reg_17, d_reg_16, d_reg_15]
```

```python
# Preview first 5 rows of regular season offense dataframes:
p.print_props(o_reg_dfs, '.head()')
```

```python
# View column features of regular season offense dataframes:
p.print_props(o_reg_dfs, '.columns')
```

```python
# Examine data types of regular season offense dataframes:
p.print_props(o_reg_dfs, '.dtypes')
```

```python
# View shape of regular season offense dataframes:
p.print_props(o_reg_dfs, '.shape')
```

```python
# Examine columns with null values of regular season offense dataframes:
p.print_props(o_reg_dfs, '.isnull().sum()')
```

```python
# Preview first 5 rows of regular season defense dataframes:
p.print_props(d_reg_dfs, '.head()')
```

```python
# Preview column features of regular season defense dataframes:
p.print_props(d_reg_dfs, '.columns')
```

```python
# Preview data types of regular season defense dataframes:
p.print_props(d_reg_dfs, '.dtypes')
```

```python
# Preview shape of regular season defense dataframes:
p.print_props(d_reg_dfs, '.shape')
```

```python
# Preview columns with null values of regular season defense dataframes:
p.print_props(d_reg_dfs, '.isnull().sum()')
```

```python
# Now that all dataframes have been verfified for use, concatenate:
o_final_df = pd.concat(o_reg_dfs, ignore_index = True)
d_final_df = pd.concat(d_reg_dfs, ignore_index = True)
```

```python
o_final_df['Year'].value_counts()
```

```python
d_final_df['Year'].value_counts()
```

```python
o_final_df.name = 'Combined Best Offensive Teams Per Season'
d_final_df.name = 'Combined Best Defensive Teams Per Season'
```

```python
combined_dfs = [o_final_df, d_final_df]
```

```python
# Preview first 5 rows of cleaned dataframes:
p.print_props(combined_dfs, '.head()')
```

```python
# Preview last 5 rows of cleaned dataframes:
p.print_props(combined_dfs, '.tail()')
```

```python
# Examine data types of cleaned dataframes:
p.print_props(combined_dfs, '.dtypes')
```

```python
o_features = [col for col in o_final_df.columns if col not in ['TeamName', 'TimeOfPossession']]
o_final_df[o_features] = o_final_df[o_features].astype(float)
d_features = [col for col in d_final_df.columns if col not in ['TeamName', 'OpponentTimeOfPossession']]
d_final_df[d_features] = d_final_df[d_features].astype(float)
```

```python
# Examine adjusted data types as floats:
p.print_props(combined_dfs, '.dtypes')
```
