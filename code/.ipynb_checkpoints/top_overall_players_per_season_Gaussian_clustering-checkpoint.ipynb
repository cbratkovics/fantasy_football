```python
# DataFrame manipulation libraries:
import pandas as pd
import numpy as np
# import props as p

# Visualization libraries:
import matplotlib.pyplot as plt
import seaborn as sns

# Gaussian clustering libraries:
from sklearn.mixture import GaussianMixture
from matplotlib import cm

%matplotlib inline
```

```python
o_df = pd.read_csv('../data/clean_files/top_overall_offensive_player_data.csv')
# o_df = pd.read_csv('../clean_data/clean_player_weekly_more_data_2018.csv')
```

```python
o_df.head()
```

```python
o_df['Position'].value_counts()
```

```python
def color_positions(pos):
    try:
        if pos == 'WR':
            return 'red'
        if pos == 'RB':
            return 'blue'
        if pos == 'TE':
            return 'orange'
        if pos == 'QB':
            return 'green'
        if pos == 'FB':
            return 'yellow'
        
        return 'black'
    
    except:
        return 'black'
```

```python
plt.figure(figsize = (20, 10))

plt.scatter(o_df['Rank'], o_df['FantasyPoints'], 
            c = o_df['Position'].apply(color_positions))

plt.title('Fantasy Points Per Season', fontsize=18)
plt.ylabel('Fantasy Points', fontsize=18)
plt.xlabel('Player Rank', fontsize=18)
plt.xticks(np.arange(0, 301, step=30), size=14)
plt.yticks(np.arange(0, 451, step=50), size=14)
plt.savefig('../images/initial_clustering_2015_2018.png')
plt.show();
```

```python
clusters = 8
g_mix = GaussianMixture(n_components = clusters, random_state = 42)
```

```python
X = o_df[['FantasyPoints']]
g_mix.fit(X)
rank_cluster = g_mix.predict(X)
```

```python
plt.figure(figsize = (20, 10))

plt.scatter(o_df['Rank'], o_df['FantasyPoints'], 
            c = [cm.Spectral(rank/clusters) for rank in rank_cluster])

plt.title('Fantasy Points Per Season; 1-Dimensional Gaussian Clusters', fontsize=18)
plt.ylabel('Fantasy Points', fontsize=18)
plt.xlabel('Player Rank', fontsize=18)
plt.xticks(np.arange(0, 301, step=30), size=14)
plt.yticks(np.arange(0, 451, step=50), size=14)
plt.savefig('../images/adjusted_Gaussian_clustering_2015_2018.png')
plt.show();
```

```python
o_df.shape
```

```python
best_rank = o_df.loc[0][['Rank', 'FantasyPoints']]
worst_rank = o_df.loc[299][['Rank', 'FantasyPoints']]
print('Best Season Rank = ', best_rank[0])
print('Best Fantasy Points In a Season = ', best_rank[1])
print()
print('Worst Season Rank = ', worst_rank[0])
print('Worst Fantasy Points In a Season = ', worst_rank[1])
```

```python
plt.figure(figsize = (20, 10))
plt.plot([best_rank[0], worst_rank[0]], [best_rank[1], worst_rank[1]], color = 'black')
plt.scatter(o_df['Rank'], o_df['FantasyPoints'], 
            c = [cm.Spectral(rank/clusters) for rank in rank_cluster])

plt.title('Fantasy Points Per Season', fontsize=18)
plt.ylabel('Fantasy Points', fontsize=18)
plt.xlabel('Player Rank', fontsize=18)
plt.xticks(np.arange(0, 301, step=30), size=14)
plt.yticks(np.arange(0, 451, step=50), size=14)

plt.savefig('../images/adjusted_Gaussian_clustering_2015_2018_line_of_best_fit.png')
plt.show();
```

```python
estimate = best_rank - worst_rank
vals = o_df.loc[:299][['Rank', 'FantasyPoints']] - worst_rank

# Scale by the player's adjusted Gaussian rank using dot product:
c = np.dot(vals, estimate) / np.dot(estimate, estimate)
```

```python
g_mix.fit(o_df[['Rank', 'FantasyPoints']])
rank_2d_cluster = g_mix.predict(o_df[['Rank', 'FantasyPoints']])
```

```python
o_df['1D_Cluster'] = rank_cluster
o_df['2D_Cluster'] = rank_2d_cluster
```

```python
plt.figure(figsize = (20, 10))

plt.scatter(o_df['Rank'], o_df['FantasyPoints'], 
            c = [cm.Spectral(rank/clusters) for rank in rank_2d_cluster])

plt.title('Fantasy Points Per Season; 2-Dimensional Gaussian Clusters', fontsize=18)
plt.ylabel('Fantasy Points', fontsize=18)
plt.xlabel('Player Rank', fontsize=18)
plt.xticks(np.arange(0, 301, step=30), size=14)
plt.yticks(np.arange(0, 451, step=50), size=14)
plt.savefig('../images/adjusted_2D_Gaussian_clustering_2015_2018.png')
plt.show();
```
