```python
# DataFrame manipulation libraries:
import pandas as pd
import numpy as np
import props as p

# Visualization libraries:
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Evaluation and processing libraries:
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Keras libraries:
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Dense, Dropout, GRU
from keras.optimizers import Adam
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
df = pd.read_csv('../clean_data/weekly_player_data_2015_to_2018_more_columns.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
df.dtypes
```

```python
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)
df.head()
```

```python
df.shape
```

```python
features = [col for col in df.columns if col not in ['Name', 'Team', 'Position', 'Opponent', 'FantasyPoints']]
df[features] = df[features].astype(float)
X = df[[col for col in df.columns if col != 'FantasyPoints']]
y = df['FantasyPoints']
```

```python
X.dtypes
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train.drop(columns = ['Name', 'Team', 'Position', 'Opponent']))
X_test_sc = ss.transform(X_test.drop(columns = ['Name', 'Team', 'Position', 'Opponent']))
```

```python
X_test_sc[0:5]
```

```python
model = Sequential()
model.add(Dense(121, activation='relu', input_shape=(11,)))
model.add(Dense(99, activation='relu'))
model.add(Dense(66, activation='relu'))
model.add(Dense(33, activation='relu'))
model.add(Dense(22, activation='relu'))
model.add(Dense(11, activation='relu'))
# model.add(Dense(2, activation='relu'))
model.add(Dense(1))
```

```python
# early_stop = EarlyStopping(monitor='val_loss', min_delta=0, patience=15, verbose=1, mode='auto')
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```

```python
history = model.fit(X_train_sc, y_train, validation_data=(X_test_sc, y_test), epochs=120, 
                    batch_size=256) #, callbacks = [early_stop])
```

```python
preds = model.predict(X_test_sc)
preds
```

```python
y_test.head()
```

```python
train_loss = history.history['loss']
test_loss = history.history['val_loss']

plt.figure(figsize=(15, 10))

# Plot training and testing loss for the original model:
plt.plot(train_loss, label='Training loss', color='navy')
plt.plot(test_loss, label='Testing loss', color='skyblue')

plt.legend();

plt.savefig('../images/adjusted_FFNN_fantasy_predictions_loss_function_plot.png')
```
