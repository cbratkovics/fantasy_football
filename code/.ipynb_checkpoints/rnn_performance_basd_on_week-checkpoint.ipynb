```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
# X_train_sc = ss.fit_transform(X_train.drop(columns = ['Name', 'Position', 'Opponent']))
# X_test_sc = ss.transform(X_test.drop(columns = ['Name', 'Position', 'Opponent']))
X_train_sc = ss.fit_transform(X_train.drop(columns = ['Name']))
X_test_sc = ss.transform((X_test.drop(columns = ['Name'])))
```

```python
test_sequences = TimeseriesGenerator(X_test_sc, y_test, length=4, batch_size=64)
```

```python
# Testing sequences of the first batch.
test_sequences[0][0].shape # first X
```

```python
test_sequences[0][1].shape # first y
```

```python
model = Sequential()
model.add(GRU(76, input_shape=(4, 76), return_sequences=True))
model.add(GRU(76, return_sequences = False)) 
model.add(Dense(38, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(lr=.001), 
              loss='binary_crossentropy', 
              metrics=['acc'])

# fit_generator objects have their 
history = model.fit_generator(train_sequences,
                    epochs = 10,
                    validation_data = test_sequences,
                    verbose = 1)
```
