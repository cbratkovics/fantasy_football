```python
import pandas as pd
import numpy as np
import props as p
%matplotlib inline
```

```python
# Top NFL offensive players 2010 - 2018.

# Passing:
o_pass_18 = pd.read_csv('../data/top_nfl_players/offense/passing/o_passing_18.csv')
o_pass_17 = pd.read_csv('../data/top_nfl_players/offense/passing/o_passing_17.csv')
o_pass_16 = pd.read_csv('../data/top_nfl_players/offense/passing/o_passing_16.csv')
o_pass_15 = pd.read_csv('../data/top_nfl_players/offense/passing/o_passing_15.csv')

# Receiving:
o_rec_18 = pd.read_csv('../data/top_nfl_players/offense/receiving/o_receiving_18.csv')
o_rec_17 = pd.read_csv('../data/top_nfl_players/offense/receiving/o_receiving_17.csv')
o_rec_16 = pd.read_csv('../data/top_nfl_players/offense/receiving/o_receiving_16.csv')
o_rec_15 = pd.read_csv('../data/top_nfl_players/offense/receiving/o_receiving_15.csv')

# Rushing:
o_rush_18 = pd.read_csv('../data/top_nfl_players/offense/rushing/o_rush_18.csv')
o_rush_17 = pd.read_csv('../data/top_nfl_players/offense/rushing/o_rush_17.csv')
o_rush_16 = pd.read_csv('../data/top_nfl_players/offense/rushing/o_rush_16.csv')
o_rush_15 = pd.read_csv('../data/top_nfl_players/offense/rushing/o_rush_15.csv')
```

```python
# Add a new column to represent the year associated with each player's performance.

# Top offensive player years.

# Passing:
o_pass_18['Year'] = 2018
o_pass_17['Year'] = 2017
o_pass_16['Year'] = 2016
o_pass_15['Year'] = 2015

# Receiving:
o_rec_18['Year'] = 2018
o_rec_17['Year'] = 2017
o_rec_16['Year'] = 2016
o_rec_15['Year'] = 2015

# Rushing:
o_rush_18['Year'] = 2018
o_rush_17['Year'] = 2017
o_rush_16['Year'] = 2016
o_rush_15['Year'] = 2015
```

```python
# Assignment of names to all dataframes.

# Passing:
o_pass_18.name = 'Top NFL Passers of 2018'
o_pass_17.name = 'Top NFL Passers of 2017'
o_pass_16.name = 'Top NFL Passers of 2016'
o_pass_15.name = 'Top NFL Passers of 2015'

# Receiving:
o_rec_18.name = 'Top NFL Receivers of 2018'
o_rec_17.name = 'Top NFL Receivers of 2017'
o_rec_16.name = 'Top NFL Receivers of 2016'
o_rec_15.name = 'Top NFL Receivers of 2015'

# Rushing:
o_rush_18.name = 'Top NFL Rushers of 2018'
o_rush_17.name = 'Top NFL Rushers of 2017'
o_rush_16.name = 'Top NFL Rushers of 2016'
o_rush_15.name = 'Top NFL Rushers of 2015'
```

```python
# Add all regular season offense and defense dataframes to a list to allow use of print_props funtion.

# Top passing players:
o_pass_dfs = [o_pass_18, o_pass_17, o_pass_16, o_pass_15]

# Top receiving players:
o_rec_dfs = [o_rec_18, o_rec_17, o_rec_16, o_rec_15]

# Top rushing players:
o_rush_dfs = [o_rush_18, o_rush_17, o_rush_16, o_rush_15]
```

```python
# Add a column to each dataframe to distinguish performance in passing, receiving, and rushing.
for ps in o_pass_dfs:
    ps['Category'] = 'Passing'
    
for rec in o_rec_dfs:
    rec['Category'] = 'Receiving'
    
for rush in o_rush_dfs:
    rush['Category'] = 'Rushing'
```

```python
# Now that all dataframes have been verfified for use, concatenate:
o_pass_final_df = pd.concat(o_pass_dfs, ignore_index = True)
o_rec_final_df = pd.concat(o_rec_dfs, ignore_index = True)
o_rush_final_df = pd.concat(o_rush_dfs, ignore_index = True)
```

```python
o_pass_final_df['Year'].value_counts()
```

```python
o_rec_final_df['Year'].value_counts()
```

```python
o_rush_final_df['Year'].value_counts()
```

```python
o_pass_final_df.name = 'Top Passers Per Season'
o_rec_final_df.name = 'Top Receivers Per Season'
o_rush_final_df.name = 'Top Rushers Per Season'
```

```python
clean_dfs = [o_pass_final_df, o_rec_final_df, o_rush_final_df]
```

```python
# Preview first 5 rows of cleaned dataframes:
p.print_props(clean_dfs, '.head()')
```

```python
# Preview last 5 rows of cleaned dataframes:
p.print_props(clean_dfs, '.tail()')
```

```python
# Examine data types of cleaned dataframes:
p.print_props(clean_dfs, '.dtypes')
```

```python
pass_features = [col for col in o_pass_final_df.columns if col not in ['Name', 'Team', 'Position', 'Category']]
o_pass_final_df[pass_features] = o_pass_final_df[pass_features].astype(float)

rec_features = [col for col in o_rec_final_df.columns if col not in ['Name', 'Team', 'Position', 'Category']]
o_rec_final_df[rec_features] = o_rec_final_df[rec_features].astype(float)

rush_features = [col for col in o_rush_final_df.columns if col not in ['Name', 'Team', 'Position', 'Category']]
o_rush_final_df[rush_features] = o_rush_final_df[rush_features].astype(float)
```

```python
# Examine data types of cleaned dataframes:
p.print_props(clean_dfs, '.dtypes')
```

```python
# Examine columns with null values of cleaned dataframes:
p.print_props(clean_dfs, '.isnull().sum()')
```

```python
# Concatenated dataframes have been verified for use, save to csv:
o_pass_final_df.to_csv('../clean_data/top_passing_players_data.csv', index = False)
o_rec_final_df.to_csv('../clean_data/top_receiving_players_data.csv', index = False)
o_rush_final_df.to_csv('../clean_data/top_rushing_players_data.csv', index = False)
```
