```python
# Import libraries:
import pandas as pd
import numpy as np
import props as p
```

```python
# Overall NFL regular season offensive players 2010 - 2018:

o_players_18 = pd.read_csv('../../raw_data/top_nfl_players/overall/offense/o_stat_best_18.csv')
o_players_17 = pd.read_csv('../../raw_data/top_nfl_players/overall/offense/o_stat_best_17.csv')
o_players_16 = pd.read_csv('../../raw_data/top_nfl_players/overall/offense/o_stat_best_16.csv')
o_players_15 = pd.read_csv('../../raw_data/top_nfl_players/overall/offense/o_stat_best_15.csv')
```

```python
# Overall NFL regular season defensive players 2010 - 2018:

d_players_18 = pd.read_csv('../../raw_data/top_nfl_players/overall/defense/d_stat_best_18.csv')
d_players_17 = pd.read_csv('../../raw_data/top_nfl_players/overall/defense/d_stat_best_17.csv')
d_players_16 = pd.read_csv('../../raw_data/top_nfl_players/overall/defense/d_stat_best_16.csv')
d_players_15 = pd.read_csv('../../raw_data/top_nfl_players/overall/defense/d_stat_best_15.csv')
```

```python
# Add a new column to represent the year associated with each player's performance.

# Years associated with overall best offensive player performances:
o_players_18['Year'] = 2018
o_players_17['Year'] = 2017
o_players_16['Year'] = 2016
o_players_15['Year'] = 2015

# Years associated with overall best defensive player performances:
d_players_18['Year'] = 2018
d_players_17['Year'] = 2017
d_players_16['Year'] = 2016
d_players_15['Year'] = 2015
```

```python
# Assignment of names to all dataframes.

# Regular season offense dataframes:
o_players_18.name = 'Top Offensive Players 2018'
o_players_17.name = 'Top Offensive Players 2017'
o_players_16.name = 'Top Offensive Players 2016'
o_players_15.name = 'Top Offensive Players 2015'

# Regular season defense dataframes:
d_players_18.name = 'Top Defensive Players 2018'
d_players_17.name = 'Top Defensive Players 2017'
d_players_16.name = 'Top Defensive Players 2016'
d_players_15.name = 'Top Defensive Players 2015'
```

```python
# Add all offensive and defensive player dataframes to a list to allow usage of print_props funtion.

# Top offensive players:
o_player_dfs = [o_players_18, o_players_17, o_players_16, o_players_15]

# Top defensive players:
d_player_dfs = [d_players_18, d_players_17, d_players_16, d_players_15]
```

```python
# Preview first 5 rows of top offensive player dataframes:
p.print_props(o_player_dfs, '.head()')
```

```python
# Examine data types of top offensive player dataframes:
p.print_props(o_player_dfs, '.dtypes')
```

```python
# View column features of top offensive player dataframes:
p.print_props(o_player_dfs, '.columns')
```

```python
# View shape of regular top offensive player dataframes:
p.print_props(o_player_dfs, '.shape')
```

```python
# Examine columns with null values of top offensive player dataframes:
p.print_props(o_player_dfs, '.isnull().sum()')
```

```python
# Preview first 5 rows of top defensive player dataframes:
p.print_props(d_player_dfs, '.head()')
```

```python
# Examine data types of top defensive player dataframes:
p.print_props(d_player_dfs, '.dtypes')
```

```python
# View column features of top defensive player dataframes:
p.print_props(d_player_dfs, '.columns')
```

```python
# View shape of regular top defensive player dataframes:
p.print_props(d_player_dfs, '.shape')
```

```python
# Examine columns with null values of top defensive player dataframes:
p.print_props(d_player_dfs, '.isnull().sum()')
```

```python
# Save overall offense and defense dataframes per season with new 'Year' feature as CSV files:

# Offense:
o_players_18.to_csv('../../clean_data/best_o_players_2018_season.csv', index=False)
o_players_17.to_csv('../../clean_data/best_o_players_2017_season.csv', index=False)
o_players_16.to_csv('../../clean_data/best_o_players_2016_season.csv', index=False)
o_players_15.to_csv('../../clean_data/best_o_players_2015_season.csv', index=False)

# Defense:
d_players_18.to_csv('../../clean_data/best_d_players_2018_season.csv', index=False)
d_players_17.to_csv('../../clean_data/best_d_players_2017_season.csv', index=False)
d_players_16.to_csv('../../clean_data/best_d_players_2016_season.csv', index=False)
d_players_15.to_csv('../../clean_data/best_d_players_2015_season.csv', index=False)
```

```python
# Now that all dataframes have been verfified for use, concatenate:
o_players_final_df = pd.concat(o_player_dfs, ignore_index = True)
d_players_final_df = pd.concat(d_player_dfs, ignore_index = True)
```

```python
o_players_final_df['Year'].value_counts()
```

```python
d_players_final_df['Year'].value_counts()
```

```python
o_players_final_df.name = 'Combined Data for Top Offensive Players'
d_players_final_df.name = 'Combined Data for Top Defensive Players'
combined_dfs = [o_players_final_df, d_players_final_df]
```

```python
# Preview first 5 rows of combined dataframes:
p.print_props(combined_dfs, '.head()')
```

```python
# Preview last 5 rows of combined dataframes:
p.print_props(combined_dfs, '.tail()')
```

```python
# Examine data types of combined dataframes:
p.print_props(combined_dfs, '.dtypes')
```

```python
# View column features of combined dataframes:
p.print_props(combined_dfs, '.columns')
```

```python
# View shape of combined dataframes:
p.print_props(combined_dfs, '.shape')
```

```python
# Examine columns with null values of combined dataframes:
p.print_props(combined_dfs, '.isnull().sum()')
```

```python
# Concatenated dataframes have been verified for use, save to csv:
o_players_final_df.to_csv('../../clean_data/top_overall_offensive_player_data.csv', index=False)
d_players_final_df.to_csv('../../clean_data/top_overall_defensive_player_data.csv', index=False)
```
