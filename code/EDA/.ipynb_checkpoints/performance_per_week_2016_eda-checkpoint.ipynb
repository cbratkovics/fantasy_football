```python
weeks_16 = pd.date_range(start='2016-09-08', end='2017-01-01', freq = 'W')
weeks_16
```

```python
count = 0
date_list = []
for i in range(1, 18):
    cond = top_weekly_16.loc[top_weekly_16['Week'] == i]
    for index, row in cond.iterrows():
        date_list.append(weeks_16[count])
    count += 1
```

```python
len(date_list)
```

```python
top_weekly_16['Date'] = [i for i in date_list]
top_weekly_16['Date'].value_counts()
```

```python
top_weekly_16['Date'] = pd.to_datetime(top_weekly_16['Date'], format = '%Y-%m-%d')
top_weekly_16['Year'] = 2016
top_weekly_16.drop(columns = ['FantasyPointsPerGame'], inplace=True)
top_weekly_16.dtypes
```
