```python
weeks_15 = pd.date_range(start='2015-09-10', end='2016-01-03', freq = 'W')
weeks_15
```

```python
count = 0
date_list = []
for i in range(1, 18):
    cond = top_weekly_15.loc[top_weekly_15['Week'] == i]
    for index, row in cond.iterrows():
        date_list.append(weeks_15[count])
    count += 1
```

```python
len(date_list)
```

```python
top_weekly_15['Date'] = [i for i in date_list]
top_weekly_15['Date'].value_counts()
```

```python
top_weekly_15['Date'] = pd.to_datetime(top_weekly_15['Date'], format = '%Y-%m-%d')
top_weekly_15['Year'] = 2015
top_weekly_15.drop(columns = ['FantasyPointsPerGame'], inplace=True)
top_weekly_15.dtypes
```
