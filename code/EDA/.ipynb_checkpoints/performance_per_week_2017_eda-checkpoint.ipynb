```python
weeks_17 = pd.date_range(start='2017-09-07', end='2017-12-31', freq = 'W')
weeks_17
```

```python
count = 0
date_list = []
for i in range(1, 18):
    cond = top_weekly_17.loc[top_weekly_17['Week'] == i]
    for index, row in cond.iterrows():
        date_list.append(weeks_17[count])
    count += 1
```

```python
len(date_list)
```

```python
top_weekly_17['Date'] = [i for i in date_list]
top_weekly_17['Date'].value_counts()
```

```python
top_weekly_17['Date'] = pd.to_datetime(top_weekly_17['Date'], format = '%Y-%m-%d')
top_weekly_17['Year'] = 2017
top_weekly_17.drop(columns = ['FantasyPointsPerGame'], inplace=True)
top_weekly_17.dtypes
```
